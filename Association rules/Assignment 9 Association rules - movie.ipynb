# Lets see the association rules

support 10% confidence > 70%

```python
freq_item=apriori(movies_new,min_support=0.1,use_colnames=True)
freq_item
```

```python
rules=association_rules(freq_item,metric='lift',min_threshold=0.70)
rules
```

```python
rules[(rules.lift>1)& (rules.confidence>0.50)]
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[['support','confidence']].hist()
```

```python
plt.scatter(rules.support,rules.confidence)
```
